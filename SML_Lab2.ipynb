```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_data.csv')
```

```python
months = df['month_number']
profit = df['total_profit']

#plot the graph
plt.plot(months,profit)
#display the graph
plt.show()
```

```python
#plot the graph
plt.plot(months,profit,color="red",label="profit", linewidth=3)
#title
plt.title("Month-wise Total Profit")
#Add labels
plt.xlabel("Months")
plt.ylabel("Total Profit")
#Add legend after plot
plt.legend(loc="lower right")
#display the graph
plt.show()
```

```python
shirts = df['shirts']
tshirts = df['t shirts']
shoes = df['shoes']
trousers = df['trousers']


plt.plot(months, shirts, label='Shirts', color='blue', linewidth=3, marker='o')
plt.plot(months, tshirts, label='T-Shirts', color='green', linewidth=3, marker='o')
plt.plot(months, trousers, label='Trousers', color='red', linewidth=3, marker='o')
plt.plot(months, shoes, label='Shoes', color='purple', linewidth=3, marker='o')

plt.title('Monthly Sales Data')
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend(loc='upper right')

plt.show()
```
